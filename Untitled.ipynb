```python
import geopandas as gpd
import pandas    as pd
import numpy     as np
import glob

shape_file = 'ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp'
daily_rpts = 'csse_covid_19_data/csse_covid_19_daily_reports/*.csv'
countrycsv = 'csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv'

map_df = gpd.read_file(shape_file)

cty_df = pd.read_csv(countrycsv)

rpt_df = pd.DataFrame()

for f in glob.glob(daily_rpts):
    df = pd.read_csv(f)
    rpt_df = rpt_df.append(df,ignore_index=True)
```

```python
map_df.plot(figsize=(15, 13))
```

```python
mrg_cty_df = cty_df.merge( rpt_df
                         , left_on  = 'Country_Region'
                         , right_on = 'Country/Region'
                         )
```

```python
mrg_map_df = map_df.merge( mrg_cty_df
                         , left_on  = 'ADM0_A3'
                         , right_on = 'iso3'
                         )
```
